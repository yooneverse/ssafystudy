```python
# ===============================
# SSAFY AI ì›”ë§í‰ê°€ ì‹¤ì „ ëŒ€ë¹„ (ì—°ìŠµìš© í…œí”Œë¦¿)
# ë²”ìœ„: EDA Â· ì„ í˜•íšŒê·€ Â· RAG ê¸°ë³¸
# ===============================

# -------------------------------
# ğŸ§© [ë¬¸ì œ 1] EDA â€“ ë°ì´í„° íƒìƒ‰ ë° ê²°ì¸¡ì¹˜ ì²˜ë¦¬
# -------------------------------

import pandas as pd

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv("student.csv")

# (1) ê²°ì¸¡ì¹˜ í˜„í™© í™•ì¸
print("ê²°ì¸¡ì¹˜ ê°œìˆ˜ í™•ì¸")
print(df.isnull().sum())          # (a) ê²°ì¸¡ì¹˜ í™•ì¸ í•¨ìˆ˜

# (2) ê²°ì„íšŸìˆ˜ì˜ í‰ê· ê°’ìœ¼ë¡œ ê²°ì¸¡ì¹˜ ì±„ìš°ê¸°
df['ê²°ì„íšŸìˆ˜'] = df['ê²°ì„íšŸìˆ˜'].fillna(df['ê²°ì„íšŸìˆ˜'].mean)  # (b) fillna í•¨ìˆ˜ ì‚¬ìš©

# (3) ê²°ì¸¡ì¹˜ ì²˜ë¦¬ í›„ ê¸°ì´ˆ í†µê³„ëŸ‰ í™•ì¸
print("ê¸°ì´ˆ í†µê³„ëŸ‰")
print(df.describe())               # (c) describe í•¨ìˆ˜ë¡œ ìš”ì•½ ì¶œë ¥


# -------------------------------
# ğŸ§© [ë¬¸ì œ 2] ì„ í˜•íšŒê·€ â€“ ê³µë¶€ì‹œê°„ê³¼ ì„±ì ì˜ ê´€ê³„
# -------------------------------

from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv("study.csv")

X = df[['ê³µë¶€ì‹œê°„']]
y = df['ì„±ì ']

# (1) ëª¨ë¸ í•™ìŠµ
model = LinearRegression()
model.fit(X, y)

# (2) ì˜ˆì¸¡
y_pred = model.predict(X)

# (3) ì‹œê°í™”
plt.scatter(X, y, label='ì‹¤ì œê°’')
plt.____(X, y_pred, color='red', label='ì˜ˆì¸¡ì„ ')  # (a) plot í•¨ìˆ˜
plt.xlabel('ê³µë¶€ì‹œê°„')
plt.ylabel('ì„±ì ')
plt.legend()
plt.show()

# (4) ëª¨ë¸ ê³„ìˆ˜ ë° ì ˆí¸ ì¶œë ¥
print("ê¸°ìš¸ê¸°:", model.coef_)
print("ì ˆí¸:", model.intercept_)

# (5) ì„œìˆ í˜• ì—°ìŠµ
# - coef_: ê³µë¶€ì‹œê°„ì´ 1 ì¦ê°€í•  ë•Œ ì„±ì ì´ ì–¼ë§ˆë‚˜ ì˜¤ë¥´ëŠ”ê°€
# - intercept_: ê³µë¶€ì‹œê°„ì´ 0ì¼ ë•Œì˜ ì˜ˆì¸¡ ì„±ì 


# -------------------------------
# ğŸ§© [ë¬¸ì œ 3] RAG â€“ ë¬¸ì„œ ê²€ìƒ‰ íŒŒì´í”„ë¼ì¸ ê¸°ë³¸
# -------------------------------

from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

texts = ["AI is the future", "I love studying AI", "Machine Learning is part of AI"]

# (1) ì„ë² ë”©
emb = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# (2) ë²¡í„°DB ìƒì„±
db = Chroma.from_texts(texts, embedding=emb)

# (3) ê²€ìƒ‰
query = "What is AI?"
result = db.____(query)        # (a) similarity_search ë©”ì„œë“œ

print("ê²€ìƒ‰ ê²°ê³¼:")
print(result[0].page_content)
```
